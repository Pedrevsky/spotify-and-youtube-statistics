```python
from os import environ, chdir
import pandas as pd
import psycopg2 as ps
```

```python
dbname = "spotify_and_yt"
port = 5433
username = "postgres"
sql_password = environ.get("postgres_passwd")
```

```python
conn = ps.connect(database=dbname, port=port, user=username, password=sql_password)
```

```python
curr = conn.cursor()
```

```python
curr.execute("SELECT * FROM songs_stats")
pd.DataFrame(curr.fetchall())
```

```python
curr.execute("SELECT * FROM songs_info")
songs_info = pd.DataFrame(curr.fetchall(), columns=["album_id", "song_id", "title", "duration_ms", "artists"])
songs_info
```

```python
curr.execute("SELECT * FROM songs_stats")
songs_stats = pd.DataFrame(curr.fetchall(), columns=["song_id", "playcount", "date"])
songs_stats
```

```python
dff = songs_info.merge(songs_stats, on="song_id")
```
